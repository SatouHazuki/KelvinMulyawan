```python
import pandas as pd
survey_data = pd.read_csv('../young-people-survey-responses.csv')
survey_data.head()
```

```python
survey_data.drop('Unnamed: 0',1,inplace=True)
survey_data.head()
```

```python
survey_data.isnull().sum()
```

```python
survey_data['Age'] = survey_data['Age'].fillna(survey_data['Age'].mean())
survey_data['Age'].isnull().sum()
```

```python
survey_data.info()
```

```python
survey_data['Age'] = survey_data['Age'].astype('int')
```

```python
ac = []
for age in survey_data['Age']:
    if age < 21:
        ac.append('Less Than 21')
    else:
        ac.append('21+')
survey_data['Age Category'] = ac
survey_data.head()
```
